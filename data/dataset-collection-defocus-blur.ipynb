## 4. Defocus-blur Images

```python
import os
import shutil

def merge_files(src_path, dest_path):
    """Gộp tất cả các tệp từ src_path vào dest_path"""
    os.makedirs(dest_path, exist_ok=True)
    
    # Dictionary để theo dõi số lần xuất hiện của mỗi tên file
    file_count = {}
    
    for root, dirs, files in os.walk(src_path):
        for file in files:
            src_file_path = os.path.join(root, file)
            
            # Lấy tên file (không bao gồm phần đuôi)
            file_name, file_extension = os.path.splitext(file)
            
            # Kiểm tra nếu tên file đã xuất hiện trước đó
            if file_name in file_count:
                file_count[file_name] += 1
                new_file_name = f"{file_name}({file_count[file_name]}){file_extension}"
            else:
                file_count[file_name] = 0
                new_file_name = f"{file_name}{file_extension}"
            
            dest_file_path = os.path.join(dest_path, new_file_name)
            
            # Kiểm tra xem tệp đã tồn tại chưa
            while os.path.exists(dest_file_path):
                file_count[file_name] += 1
                new_file_name = f"{file_name}({file_count[file_name]}){file_extension}"
                dest_file_path = os.path.join(dest_path, new_file_name)
            
            print(f"Copied: {file} --> {dest_path}/{new_file_name}")
            shutil.copy(src_file_path, dest_file_path)
            
# Đường dẫn đến thư mục chứa noisy images
trainc = "/kaggle/input/defocus-blur-train/train/inputC"
trainl = "/kaggle/input/defocus-blur-train/train/inputL"
trainr = "/kaggle/input/defocus-blur-train/train/inputR"
testc = "/kaggle/input/defocus-blur-test-val/test/test/inputC"
testl = "/kaggle/input/defocus-blur-test-val/test/test/inputL"
testr = "/kaggle/input/defocus-blur-test-val/test/test/inputR"
valc = "/kaggle/input/defocus-blur-test-val/val/val/inputC"
vall = "/kaggle/input/defocus-blur-test-val/val/val/inputL"
valr = "/kaggle/input/defocus-blur-test-val/val/val/inputR"

# Đường dẫn đến thư mục mới để lưu tất cả các tệp đã gộp
merged_output_path = "/kaggle/working/"

# Gộp tất cả các tệp từ các thư mục noisy images vào thư mục MERGED_images
merge_files(trainc, merged_output_path)
merge_files(trainl, merged_output_path)
merge_files(trainr, merged_output_path)
merge_files(testc, merged_output_path)
merge_files(testl, merged_output_path)
merge_files(testr, merged_output_path)
merge_files(valc, merged_output_path)
merge_files(vall, merged_output_path)
merge_files(valr, merged_output_path)
print("done")
```

```python
def rename_files_in_folder(folder_path):
    """Đổi tên tất cả các tệp trong folder thành dạng số đếm_noise.jpg"""
    count = 0
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            src_file_path = os.path.join(root, file)
            dest_file_name = f"{count:04d}_defocus-blur.jpg"
            dest_file_path = os.path.join(root, dest_file_name)
            os.rename(src_file_path, dest_file_path)
            print(f"Renamed: {file} --> {dest_file_name}")
            count += 1

# Đổi tên tất cả các tệp trong thư mục MERGED_images
rename_files_in_folder(merged_output_path)
print("done")
```
