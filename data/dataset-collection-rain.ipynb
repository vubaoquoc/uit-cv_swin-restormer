```python
import os
import shutil

# Đường dẫn đến thư mục cần xóa
working_directory = "/kaggle/working/"

# Hàm xóa tất cả các tệp và thư mục trong đường dẫn cụ thể
def delete_contents(directory):
    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)
        if os.path.isfile(item_path):
            os.remove(item_path)
            print(f"Deleted file: {item_path}")
        elif os.path.isdir(item_path):
            shutil.rmtree(item_path)
            print(f"Deleted directory: {item_path}")

# Gọi hàm để xóa nội dung trong thư mục /kaggle/working/
delete_contents(working_directory)

print("All files and folders in /kaggle/working/ have been deleted.")
```

## 2. Rain Images

```python
import os
import shutil

def copy_and_rename_rain_images(src_path, dest_path):
    """Copy và rename tất cả các tệp từ src_path vào dest_path"""
    os.makedirs(dest_path, exist_ok=True)
    count = 0
    for root, dirs, files in os.walk(src_path):
        for file in files:
            src_file_path = os.path.join(root, file)
            dest_file_name = f"{count:05d}_rain.jpg"
            dest_file_path = os.path.join(dest_path, dest_file_name)
            shutil.copy(src_file_path, dest_file_path)
            print(f"Copied: {file} --> {dest_path}/{dest_file_name}")
            count += 1

# Đường dẫn đến thư mục chứa tệp rain images
rain_input_path = "/kaggle/input/synthetic-rain/rain_dataset-20230419T131346Z-002/rain_dataset/train/train/Rain13K/input"

# Đường dẫn đến thư mục mới để lưu tất cả các tệp đã copy và rename
rain_output_path = "/kaggle/working/"

# Copy và rename tất cả các tệp từ thư mục rain input vào thư mục rain output
copy_and_rename_rain_images(rain_input_path, rain_output_path)
print("done")
```
